# PyCity Schools Analysis

Summary:
This challenge involved the construction of various data frames that compared the passing rates in math and reading in different schools.  The data consisted of many students from a total of 15 schools and what their math and reading scores are.  The size and budget of each school have been given as well, which could be used to see if there is a possible correlation between the passing rate depending on the size of the school or how much of the budget can be divided to give each student the proper support they need.  Using these pieces of data for each student, multiple data frames can be made comparing what kind of schools tend to do better, how much money is allocated to each student based on the total budget of the school, and if the size of the school has a real effect on the passing rate.  The conclusions made are the culmination of combining different pieces of data and seeing how the pass rate is affected by them.

Conclusion 1:
Based on the type_summary_df, or the database that compares the averages and passing rates depending on what type of school the students attend, charter schools have higher passing rates and averages compared to those of district schools.  The overall averages for math and reading scores are greater for students who attend charter schools.  In addition, the percentages for passing math, reading, and both subjects together are greater in charter schools rather than in district schools.  There are 8 charter schools out of the 15 total schools used in the data, and even though this means more than half of the schools are charter schools, the scores far exceed those from district schools.  No argument can be made that the scores are higher in charter schools because there are more schools in the data since the data frames used information comparing the percentages and averages of each school individually.  When comparing the charter schools individually to the district schools, the averages and passing rates still are superior in charter schools.  Even when separating the highest-performing schools from the bottom-performing schools, based on the percent of the overall passing rate, the top 5 schools that performed well were all charter schools; the top 5 schools that performed the poorest were all district schools.

Conclusion 2:
The total budget of the schools and the amount allocated to each student does not play a significant role in the passing rates of the students.  A common thinking is that if a school has enough money to offer the students they should be able to pass with higher passing rates since they have the chance to use more resources.  The sizes of the schools determine how much money would be given to them.  Schools with more students received a vastly higher budget compared to schools with fewer students, but the amount of money per student was within close range of each other.  When examining how much money can be allocated per student, there is no drastic difference between budgets per capita between schools.  Every single school has a budget per capita that is in the range of $580 to $655, and even though this seems like a big difference, this range did not have a massive impact on helping the students do better.  If anything, the data has shown that if a school has a higher budget, the overall passing rate of that school is significantly lower than that of a school with a smaller budget.  If a school has more money, this does not mean that it plays a significant task in helping students succeed, quite the opposite.  Based on this data, district schools tend to receive more money than charter schools, and when comparing any single district school to a charter school, the charter school will have a higher overall passing rate.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_csv = Path("../Resources/schools_complete.csv")
student_csv = Path("../Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_csv)
student_data_df = pd.read_csv(student_csv)

# Combine the data into a single dataset.
school_data_combined_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_combined_df.head()
```

## District Summary

```python
# Calculate the total number of unique schools

school_count = len(school_data_df["school_name"].unique())

school_count
```

```python
# Calculate the total number of students

student_count = student_data_df["student_name"].count()

student_count
```

```python
# Calculate the total budget

total_budget = school_data_df["budget"].sum()

total_budget
```

```python
# Calculate the average (mean) math score

average_math_score = student_data_df["math_score"].mean()

average_math_score
```

```python
# Calculate the average (mean) reading score

average_reading_score = student_data_df["reading_score"].mean()

average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)

passing_math_count = school_data_combined_df[(school_data_combined_df["math_score"] >= 70)].count()["student_name"]

passing_math_percentage = passing_math_count / float(student_count) * 100

passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)

passing_reading_count = school_data_combined_df[(school_data_combined_df["reading_score"] >= 70)].count()["student_name"]

passing_reading_percentage = passing_reading_count / float(student_count) * 100

passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading

passing_math_reading_count = school_data_combined_df[
    (school_data_combined_df["math_score"] >= 70) & (school_data_combined_df["reading_score"] >= 70)
].count()["student_name"]

overall_passing_rate = passing_math_reading_count /  float(student_count) * 100

overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame

district_summary_df = pd.DataFrame([{
    "Total Unique Schools": school_count,
    "Total Students": student_count,
    "Total Budget": total_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": passing_math_percentage,
    "% Passing Reading": passing_reading_percentage,
    "% Overall Pass Rate": overall_passing_rate
}])

# Formatting

district_summary_df["Total Unique Schools"] = district_summary_df["Total Unique Schools"].map("${:,.0f}".format)
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:,.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:,.2f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:,.2f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:,.2f}".format)
district_summary_df["% Overall Pass Rate"] = district_summary_df["% Overall Pass Rate"].map("{:,.2f}".format)

# Display the DataFrame
    
district_summary_df
```

## School Summary

```python
# Use the code provided to select the type per school from school_data

perschool_types = school_data_df.set_index(["school_name"])["type"]

perschool_types
```

```python
# Calculate the total student count per school from school_data

# perschool_counts = len(school_data_combined_df["school_name"].unique())

perschool_student_count = school_data_combined_df["school_name"].value_counts()

# perschool_counts
perschool_student_count
```

```python
# Calculate the total school budget and per capita spending per school from school_data
# The average is needed for the budget to see what the average budget is for the whole school, which would be the same as given in the excel
# Since the values for the budget will be the same for each school, adding the mean function will make it equal

perschool_budget = (school_data_combined_df.groupby(["school_name"])['budget']).mean()

# To calculate the perschool_capita we need the amount od students in each school as well (capita = budget / student count)
# Per capita is per student

perschool_per_capita = perschool_budget / perschool_student_count

perschool_budget
perschool_per_capita
```

```python
# Calculate the average test scores per school from school_data_complete

perschool_math_average = school_data_combined_df.groupby(["school_name"])['math_score'].mean()
perschool_math_average

perschool_reading_average = school_data_combined_df.groupby(["school_name"])['reading_score'].mean()
perschool_reading_average
```

```python
# Calculate the number of students per school with math scores of 70 or higher from school_data_complete

perschool_students_passing_math = school_data_combined_df.loc[school_data_combined_df["math_score"] >= 70]
# DataFrame made, now count it up per school

perschool_pass_math_count = perschool_students_passing_math["school_name"].value_counts()

perschool_students_passing_math #dataframe
perschool_pass_math_count
```

```python
# Calculate the number of students per school with reading scores of 70 or higher from school_data_complete

perschool_students_passing_reading = school_data_combined_df.loc[school_data_combined_df["reading_score"] >= 70]
# DataFrame made, now count it up per school

perschool_pass_reading_count = perschool_students_passing_reading["school_name"].value_counts()

perschool_students_passing_reading #dataframe
perschool_pass_reading_count
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher

perschool_students_passing_mathandreading = school_data_combined_df[
    (school_data_combined_df["reading_score"] >= 70) & (school_data_combined_df["math_score"] >= 70)
]

perschool_students_passing_mathandreading = perschool_students_passing_mathandreading.groupby(["school_name"]).size()

perschool_students_passing_mathandreading
```

```python
# Use the provided code to calculate the passing rates

perschool_math_passrate = (perschool_pass_math_count / perschool_student_count) * 100

perschool_reading_passrate = (perschool_pass_reading_count / perschool_student_count) * 100

overall_passrate = (perschool_students_passing_mathandreading / perschool_student_count) * 100


perschool_math_passrate
perschool_reading_passrate
overall_passrate
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

per_school_summary_df = pd.DataFrame({
    "Type of School": perschool_types,
    "Total Students": perschool_student_count,
    "Total School Budget": perschool_budget,
    "Budget Per Student": perschool_per_capita,
    "Average Math Score": perschool_math_average,
    "Average Reading Score": perschool_reading_average,
    "% Passing Math": perschool_math_passrate,
    "% Passing Reading": perschool_reading_passrate,
    "% Overall Pass Rate": overall_passrate
})

# Formatting

per_school_summary_df["Total Students"] = per_school_summary_df["Total Students"].map("{:,.0f}".format)
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].map("${:,.2f}".format)
per_school_summary_df["Budget Per Student"] = per_school_summary_df["Budget Per Student"].map("${:,.2f}".format)
per_school_summary_df["Average Math Score"] = per_school_summary_df["Average Math Score"].map("{:,.2f}".format)
per_school_summary_df["Average Reading Score"] = per_school_summary_df["Average Reading Score"].map("{:,.2f}".format)
per_school_summary_df["% Passing Math"] = per_school_summary_df["% Passing Math"].map("{:,.2f}".format)
per_school_summary_df["% Passing Reading"] = per_school_summary_df["% Passing Reading"].map("{:,.2f}".format)
per_school_summary_df["% Overall Pass Rate"] = per_school_summary_df["% Overall Pass Rate"].map("{:,.2f}".format)

# Display the DataFrame
per_school_summary_df
```

## Highest-Performing Schools (by % Overall Passing)

## Scores by School Spending

```python
# Establish the bins

spending_bins = [0, 585, 630, 645, 680]

labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"

school_spending_df = per_school_summary_df.copy()

school_spending_df
```

```python
#  Calculate averages for the desired columns.

# The data type for the values was object, so they need to be changed into integer
school_spending_df['Average Math Score'] = school_spending_df['Average Math Score'].astype(float)
school_spending_df['Average Reading Score'] = school_spending_df['Average Reading Score'].astype(float)
school_spending_df['% Passing Math'] = school_spending_df['% Passing Math'].astype(float)
school_spending_df['% Passing Reading'] = school_spending_df['% Passing Reading'].astype(float)
school_spending_df['% Overall Pass Rate'] = school_spending_df['% Overall Pass Rate'].astype(float)


spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()

spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()

spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()

spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()

overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Pass Rate"].mean()

school_spending_df.dtypes
```
